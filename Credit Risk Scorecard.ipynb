```python
def calc_woe_iv(data, feature, target, bins=5, categorical=False):
    tmp = data.copy()

    # Numeric → binning
    if not categorical:
        tmp[feature] = pd.qcut(tmp[feature], q=bins, duplicates='drop')

    # Group by feature bins
    grouped = tmp.groupby(feature)[target].agg(['count', 'sum'])
    grouped = grouped.rename(columns={'sum': 'bad'})
    grouped['good'] = grouped['count'] - grouped['bad']

    # Avoid divide-by-zero
    grouped = grouped[(grouped['bad'] > 0) & (grouped['good'] > 0)]

    # Distribution of good & bad
    grouped['dist_good'] = grouped['good'] / grouped['good'].sum()
    grouped['dist_bad'] = grouped['bad'] / grouped['bad'].sum()

    # WOE - it computes the distribution of good vs. bad, then calculates WOE = ln(dist_good / dist_bad)
    grouped['woe'] = np.log(grouped['dist_good'] / grouped['dist_bad'])

    # IV - Finally, it computes IV = Σ((dist_good – dist_bad) × WOE)
    grouped['iv'] = (grouped['dist_good'] - grouped['dist_bad']) * grouped['woe']
    iv = grouped['iv'].sum()

    return grouped, iv
```

```python
iv_summary = {}  # Dictionary to store variable → IV mapping

for col in numeric_vars:
    _,iv = calc_woe_iv(df,col,target, bins = 5, categorical = False)
    iv_summary[col] = iv

for col in categorical_vars:
    _, iv = calc_woe_iv(df, col, target, categorical=True)
    iv_summary[col] = iv
```

```python
def calculate_iv_simple(data, feature, target, bins=5, categorical=False):
    tmp = data.copy()

    # Numeric → binning
    if not categorical:
        tmp[feature] = pd.qcut(tmp[feature], q=bins, duplicates='drop')

    # Group by bins / categories
    grouped = tmp.groupby(feature)[target].agg(['count', 'sum'])
    grouped = grouped.rename(columns={'sum': 'bad'})
    grouped['good'] = grouped['count'] - grouped['bad']

    # Distributions (add epsilon to avoid divide-by-zero)
    grouped['dist_good'] = grouped['good'] / grouped['good'].sum()
    grouped['dist_bad'] = grouped['bad'] / grouped['bad'].sum()

    # WOE (safe)
    grouped['woe'] = np.log(
        (grouped['dist_good'] + 1e-10) / (grouped['dist_bad'] + 1e-10)
    )

    # IV
    grouped['iv'] = (grouped['dist_good'] - grouped['dist_bad']) * grouped['woe']

    # Return ONLY total IV
    return grouped['iv'].sum()

def classify_iv(iv):
    """
    Classify variable predictive power based on IV value.
    """
    if iv < 0.02:
        return "Not Useful"
    elif iv < 0.1:
        return "Weak"
    elif iv < 0.3:
        return "Medium"
    elif iv < 0.5:
        return "Strong"
    else:
        return "Suspicious / Overfit"
```

```python
def is_woe_almost_monotonic(data, feature, target, bins=5, tolerance=2):
    
    """
    Check if WOE values for a feature are *almost monotonic*.
    Allows up to `tolerance` direction changes.
    Uses fewer bins to smooth noise.
    """
    tmp = data.copy()

    # Bin numeric feature
    tmp[feature] = pd.qcut(tmp[feature], q=bins, duplicates='drop')

    # WOE calculation
    grouped = tmp.groupby(feature)[target].agg(['count', 'sum'])
    grouped['bad'] = grouped['sum']
    grouped['good'] = grouped['count'] - grouped['bad']

    dist_good = grouped['good'] / grouped['good'].sum()
    dist_bad = grouped['bad'] / grouped['bad'].sum()

    woe = np.log((dist_good + 1e-10) / (dist_bad + 1e-10))

    # Check direction changes
    diffs = np.sign(np.diff(woe.values))
    direction_changes = np.sum(diffs[1:] != diffs[:-1])

    return direction_changes <= tolerance
```
